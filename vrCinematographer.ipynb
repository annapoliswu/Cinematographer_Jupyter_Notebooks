```python
avatarParts = ["head", "leftHand", "rightHand", "root"]
avatarVariableArray = []
numAvatars = 2
for j in range(numAvatars):
    avatarVariableArray.append("avatar"+str(j)+".voiceVolume")
    for part in avatarParts:
        for posRot in ["position", "rotation"]:
            for axis in ["x","y", "z","w"]:
                    avatarVariableArray.append("avatar"+str(j)+"."+part+"."+posRot+"."+axis)
print(avatarVariableArray)
```

```python
#reads files into one unformatted df
def readFiles(dataFileName, labelFileName):
    dfs = []
    fileStream = open(dataFileName ,'r')
    dataList = fileStream.read().strip('\n').split('|')

    #read list into df
    for i in range(1, len(dataList)-1):
        jObject = json.loads(dataList[i])
        df = pd.DataFrame.from_dict(jObject, orient='index').transpose()
        #print("\n", jObject["avatarData"]["avatars"][1]["headTransform"]["position"])
        dfs.append(df)

    json_df = pd.concat(dfs, keys=range(len(dataList)), axis=0).reset_index(level=1, drop=True)
    df = json_df.copy(deep=True)

    fileStream = open(labelFileName,'r')
    jObject = json.loads(fileStream.read())
    labels_df = pd.DataFrame.from_dict(jObject, orient='index').transpose()
    labels_df.rename(columns={'list': 'label'}, inplace=True)
    df["camera"] = labels_df
    return df

#set new df entry with new column name with the data's old column and row
def setRow(df, data, colName, newColName, rowNum, itemNum):
    df.at[rowNum, newColName] = data[colName][itemNum]

#with json data of a transform value, get xyzw components and set as separate columns in entry j of set_df
def setXYZ(df, data, colName, newColName, rowNum, itemNum):
    setRow(df, data, colName+".x", newColName+".x", rowNum, itemNum)
    setRow(df, data, colName+".y", newColName+".y", rowNum, itemNum)
    setRow(df, data, colName+".z", newColName+".z", rowNum, itemNum)
    setRow(df, data, colName+".w", newColName+".w", rowNum, itemNum)

def setTransform(df, data, colName, newColName, rowNum, itemNum):
    setXYZ(df, data, colName+".position", newColName+".position", rowNum, itemNum)
    setXYZ(df, data, colName+".rotation", newColName+".rotation", rowNum, itemNum)

#takes from json file to formatted, expanded df
def jsonToDf(dataFileName, labelFileName):
    df = readFiles(dataFileName, labelFileName)
    
    # expand avatars + camera data to different columns in df
    for i in range(0, len(df)):
        df.at[i, "timeSinceRecord"] = df["timeElapsed"][i]-df["timeElapsed"][0] 
        camData = df["cameraData"][i]
        cameras = pd.json_normalize(camData["cameras"])
        #print(camData)
        
        for j in range(0, len(cameras)):
            camName = "cam"+str(j)
            setTransform(df, cameras, "transform", camName, i, j)
            setRow(df, cameras, "fov", camName +".fov", i, j)

        avatarData = df["avatarData"][i]
        avatars = pd.json_normalize(avatarData["avatars"])
        for j in range(0, len(avatars)):
            avatarName = "avatar"+str(j)
            setRow(df, avatars, "voiceVolume", avatarName+".voiceVolume", i, j)
            setTransform(df, avatars, "headTransform", avatarName+".head", i, j)
            setTransform(df, avatars, "leftHandTransform", avatarName+".leftHand", i, j)
            setTransform(df, avatars, "rightHandTransform", avatarName+".rightHand", i, j)
            setTransform(df, avatars, "rootTransform", avatarName+".root", i, j)
            
    return df


#df with added history, averages, modes according to chunk size
# if clf is given, adds prediction columns
def createChunkedDf(rough_df, chunkSize, clf = None, forecastThreshold = None):
    df = rough_df.copy()
    prevCamStartTime = df["timeElapsed"][0]
    
    prevCamera, durationOfPrevCamera, modeCamera, meanCamera, varCamera = [-1 , 0, -1, -1, 0]   
    df.insert(4, "durationOfPrevCamera", durationOfPrevCamera)
    df.insert(4, "modeCamera", modeCamera)
    df.insert(4, "varCamera", varCamera)
    df.insert(4, "meanCamera", meanCamera)
    df.insert(4, "prevCamera", prevCamera)
    
    for var in avatarVariableArray:
        df[var+ ".mean"] = -1
        df[var+ ".var"] = 0
        
    for i in range(0, len(df)):  
        start = max(0, i-chunkSize)
        stop = i
        
        if i >= 1 : #at 2nd, looking back 1
            prevCamera = df["camera"][i-1]
            modeCamera = df["camera"][start : stop].mode().values[0] #[inclusive:non-inclusive]
            meanCamera = df["camera"][start : stop].mean()
            #avatar voice, posn, rot means
            meanArray = df[avatarVariableArray][start:stop].mean()
            for v in range(len(avatarVariableArray)): #for every variable name, take the mean
                df.at[i, avatarVariableArray[v]+".mean"] = meanArray[v]
        
        if i >= 2 : #at 3rd, looking back 2
            varCamera = df["camera"][start : stop].var()
            if prevCamera != df["camera"][i-2]:
                prevCamStartTime = df["timeElapsed"][i-1]
            #avatar voice, posn, rot vars
            varArray = df[avatarVariableArray][start:stop].var()
            for v in range(len(avatarVariableArray)): #for every variable name, take the var
                df.at[i, avatarVariableArray[v]+".var"] = varArray[v]

        prevCamEndTime = df["timeElapsed"][i]
        durationOfPrevCamera = prevCamEndTime - prevCamStartTime
        df.at[i,"prevCamera"], df.at[i,"durationOfPrevCamera"], df.at[i,"modeCamera"], df.at[i,"meanCamera"], df.at[i, "varCamera"] = [prevCamera, durationOfPrevCamera, modeCamera, meanCamera, varCamera]
        
        
        if(clf != None):
            X = [selectX(df, i)]
            probaArray = clf.predict_proba(X)[0]
            maxProbaArg = probaArray.argmax()
            maxProba = probaArray[maxProbaArg]
            prediction = clf.classes_[maxProbaArg]
            
            if maxProba > forecastThreshold :
                df.at[i, "camera"] = prediction
            else:
                df.at[i,"camera"] = prevCamera
                
            #print("count: " + str(i) )
            #print("cam: " + str(df.at[i,"camera"]))
            #print("mean: " + str(meanCamera))
            #print("var: " + str(varCamera) + "\n")
        
    return df

#the last camera used, duration we stayed on the last camera, 
#and variance and mode of camera labels across a set chunk of time
def forecastData(rough_df, chunkSize, clf, forecastThreshold):
    df = createChunkedDf(rough_df, chunkSize, clf, forecastThreshold)
    return df

    
#test length of time chunk
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
#specify what to not include in training, testing, and forecasting
#dropColumns = ["timeElapsed", "cameraData", "avatarData", "camera", "prevCamera"]

selectColumns = ["timeSinceRecord", "durationOfPrevCamera", "varCamera","avatar0.voiceVolume", "avatar1.voiceVolume"]
for var in avatarVariableArray:
    selectColumns.append(var + ".mean")
    selectColumns.append(var + ".var")
dropColumns = ["cameraData", "avatarData", "camera", "modeCamera", "meanCamera", "prevCamera"]

def selectColumnArray():
    return selectColumns #df.columns.drop(dropColumns)

def selectX(df, dfIndex=None):
    thisSelectColumns = selectColumnArray()
    
    if dfIndex == None:
        return df[thisSelectColumns].values
    else:
        return df.loc[dfIndex, thisSelectColumns].values

selectColumnArray()
```

```python
# from : Brendan Artley @ Medium
# https://medium.com/@BrendanArtley/matplotlib-color-gradients-21374910584b
def hex_to_RGB(hex_str):
    """ #FFFFFF -> [255,255,255]"""
    #Pass 16 to the integer function for change of base
    return [int(hex_str[i:i+2], 16) for i in range(1,6,2)]

def get_color_gradient(c1, c2, n):
    """
    Given two hex colors, returns a color gradient
    with n colors.
    """
    assert n > 1
    c1_rgb = np.array(hex_to_RGB(c1))/255
    c2_rgb = np.array(hex_to_RGB(c2))/255
    mix_pcts = [x/(n-1) for x in range(n)]
    rgb_colors = [((1-mix)*c1_rgb + (mix*c2_rgb)) for mix in mix_pcts]
    return ["#" + "".join([format(int(round(val*255)), "02x") for val in item]) for item in rgb_colors]

#researcher model, importance of features
historySize=100
model_pkl_file = "models/Mine/model_chunk"+str(historySize)+".pkl"
clf = pickle.load(open(model_pkl_file, 'rb'))

num_features = 15
features = selectColumnArray()
importances = clf.feature_importances_
indices = np.argsort(importances)

mColors = get_color_gradient('#72E1F0', '#B83A9D', num_features)
plt.subplots(figsize=(10,6))
plt.barh(range(num_features), importances[indices[-num_features:]], color=mColors, align='center', height=.6)
plt.yticks(range(num_features), [features[i] for i in indices[-num_features:]])

plt.xlabel('Relative Importance')
plt.title("Researcher Model Feature Importance")
plt.show()
```

```python
#from : DTrimarchi10 @ Github  
#https://github.com/DTrimarchi10/confusion_matrix

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def make_confusion_matrix(cf,
                          group_names=None,
                          categories='auto',
                          count=True,
                          percent=True,
                          cbar=True,
                          xyticks=True,
                          xyplotlabels=True,
                          sum_stats=True,
                          figsize=None,
                          cmap='Blues',
                          title=None):
    '''
    This function will make a pretty plot of an sklearn Confusion Matrix cm using a Seaborn heatmap visualization.

    Arguments
    ---------
    cf:            confusion matrix to be passed in

    group_names:   List of strings that represent the labels row by row to be shown in each square.

    categories:    List of strings containing the categories to be displayed on the x,y axis. Default is 'auto'

    count:         If True, show the raw number in the confusion matrix. Default is True.

    normalize:     If True, show the proportions for each category. Default is True.

    cbar:          If True, show the color bar. The cbar values are based off the values in the confusion matrix.
                   Default is True.

    xyticks:       If True, show x and y ticks. Default is True.

    xyplotlabels:  If True, show 'True Label' and 'Predicted Label' on the figure. Default is True.

    sum_stats:     If True, display summary statistics below the figure. Default is True.

    figsize:       Tuple representing the figure size. Default will be the matplotlib rcParams value.

    cmap:          Colormap of the values displayed from matplotlib.pyplot.cm. Default is 'Blues'
                   See http://matplotlib.org/examples/color/colormaps_reference.html
                   
    title:         Title for the heatmap. Default is None.

    '''


    # CODE TO GENERATE TEXT INSIDE EACH SQUARE
    blanks = ['' for i in range(cf.size)]

    if group_names and len(group_names)==cf.size:
        group_labels = ["{}\n".format(value) for value in group_names]
    else:
        group_labels = blanks

    if count:
        group_counts = ["{0:0.0f}\n".format(value) for value in cf.flatten()]
    else:
        group_counts = blanks

    if percent:
        group_percentages = ["{0:.2%}".format(value) for value in cf.flatten()/np.sum(cf)]
    else:
        group_percentages = blanks

    box_labels = [f"{v1}{v2}{v3}".strip() for v1, v2, v3 in zip(group_labels,group_counts,group_percentages)]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0],cf.shape[1])


    # CODE TO GENERATE SUMMARY STATISTICS & TEXT FOR SUMMARY STATS
    if sum_stats:
        #Accuracy is sum of diagonal divided by total observations
        accuracy  = np.trace(cf) / float(np.sum(cf))

        #if it is a binary confusion matrix, show some more stats
        if len(cf)==2:
            #Metrics for Binary Confusion Matrices
            precision = cf[1,1] / sum(cf[:,1])
            recall    = cf[1,1] / sum(cf[1,:])
            f1_score  = 2*precision*recall / (precision + recall)
            stats_text = "\n\nAccuracy={:0.3f}\nPrecision={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(
                accuracy,precision,recall,f1_score)
        else:
            stats_text = "\n\nAccuracy={:0.3f}".format(accuracy)
    else:
        stats_text = ""


    # SET FIGURE PARAMETERS ACCORDING TO OTHER ARGUMENTS
    if figsize==None:
        #Get default figure size if not set
        figsize = plt.rcParams.get('figure.figsize')

    if xyticks==False:
        #Do not show categories if xyticks is False
        categories=False


    # MAKE THE HEATMAP VISUALIZATION
    plt.figure(figsize=figsize)
    sns.heatmap(cf,annot=box_labels,fmt="",cmap=cmap,cbar=cbar,xticklabels=categories,yticklabels=categories)

    if xyplotlabels:
        plt.ylabel('True label')
        plt.xlabel('Predicted label' + stats_text)
    else:
        plt.xlabel(stats_text)
    
    if title:
        plt.title(title)
```
